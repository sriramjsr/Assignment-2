```python
import pandas as pd
import os 
import json
import mysql.connector as sql
```

```python
path1="D:/phone pe/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path1)



clms1={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_state_list:
    p_i=path1+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            A=json.load(Data)
            for z in A['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clms1['Transaction_type'].append(Name)
              clms1['Transaction_count'].append(count)
              clms1['Transaction_amount'].append(amount)
              clms1['State'].append(i)
              clms1['Year'].append(j)
              clms1['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
df_Agg_Transaction=pd.DataFrame(clms1)
```

```python
path2="D:/phone pe/pulse/data/aggregated/user/country/india/state/"
Agg_user_list=os.listdir(path2)


clms2={'State':[], 'Year':[],'Quater':[],'Brands':[], 'Transaction_count':[], 'Percentage':[]}
for i in Agg_user_list:
    p_i=path2+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            B=json.load(Data)
            
            try:
                for z in B['data']['usersByDevice']:
                  brand=z['brand']
                  count=z['count']
                  percentage=z['percentage']
                  clms2['Brands'].append(brand)
                  clms2['Transaction_count'].append(count)
                  clms2['Percentage'].append(percentage)
                  clms2['State'].append(i)
                  clms2['Year'].append(j)
                  clms2['Quater'].append(int(k.strip('.json')))
            except:
                pass
            
df_Agg_user = pd.DataFrame(clms2)
```

```python
path4="D:/phone pe/pulse/data/map/user/hover/country/india/state/"
map_user_list=os.listdir(path4)


clms4={'State':[], 'Year':[],'Quater':[],'Districts':[], 'RegisteredUsers':[], 'AppOpens':[]}

for i in map_user_list:
    p_i=path4+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            
            
            for z in D['data']['hoverData'].items():
              districts=z[0]
              registereduser=z[1]["registeredUsers"]
              appOpens = z[1]['appOpens']
              clms4['Districts'].append(districts)
              clms4['RegisteredUsers'].append(registereduser)
              clms4['AppOpens'].append(appOpens)
              clms4['State'].append(i)
              clms4['Year'].append(j)
              clms4['Quater'].append(int(k.strip('.json')))
            

df_Map_user = pd.DataFrame(clms4)
```

```python
path5="D:/phone pe/pulse/data/top/transaction/country/india/state/"
top_trans_list=os.listdir(path5)


clms5={'State':[], 'Year':[],'Quater':[],'Pincodes':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in top_trans_list:
    p_i=path5+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            E=json.load(Data)
            
            
            
            for z in E['data']['pincodes']:
              entityname=z["entityName"]
              count = z['metric']['count']
              amount = z['metric']['amount']
              clms5['Pincodes'].append(entityname)
              clms5['Transaction_count'].append(count)
              clms5['Transaction_amount'].append(amount)
              clms5['State'].append(i)
              clms5['Year'].append(j)
              clms5['Quater'].append(int(k.strip('.json')))
            

df_Top_transac = pd.DataFrame(clms5)
```
